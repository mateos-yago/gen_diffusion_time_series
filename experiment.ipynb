```python
from time_series_generator import TimeSeriesGenerator
from time_series_ddpm_model import TimeSeriesDDPM
import network_architectures
from experiment import Experiment
import utils
```

1. Generate training data

```python
ar_data=TimeSeriesGenerator.generate_ar_series(100, 4, 0.3, burnin=4)
arma_data=TimeSeriesGenerator.generate_arma_series(100, 4, 0.2, 0.4, burnin=4)
garch_data=TimeSeriesGenerator.generate_garch_11_series(100, 4, 0.1, 0.4, 0.3, burnin=4)
```

```python
ts_models={'AR': ar_data, 'ARMA': arma_data, 'GARCH': garch_data}
```

2. Create network architectures

```python
custom_lstm_net = network_architectures.LSTMNet(input_dim=1, hidden_dim=64, num_layers=1)
custom_gru_net = network_architectures.GRUNet(input_dim=1, hidden_dim=64, num_layers=1)
custom_CNN_net = network_architectures.CNNNet(input_dim=1, hidden_dims=[32, 64], kernel_size=3, padding=1)
custom_FCNN_net = network_architectures.FCNNNet(input_dim=1, hidden_dims=[64,128,64])
custom_HybridCNNLSTM_net = network_architectures.HybridCNNLSTMNet(input_dim=1, conv_channels=16, lstm_hidden_dim=32, num_lstm_layers=2, kernel_size=3, padding=1)
```

```python
architectures={'lstm': custom_lstm_net,
              'gru': custom_gru_net,
              'CNN': custom_CNN_net,
              'FCNN': custom_FCNN_net,
              'hybridCNNLSTM': custom_HybridCNNLSTM_net}
```

```python
for ts_model in ts_models:
    for architecture in architectures:
        print(f'Working on experiment for {ts_model} model and {architecture} architecture')
        exp_name=ts_model+'_'+architecture
        experiment = Experiment(name=exp_name,
                               experiment_type=ts_model,
                               directory_path=utils.build_path('output', exp_name),
                               training_data=ts_models[ts_model],
                               network_architecture=architectures[architecture])
        experiment.execute(epochs=2)
        experiment.export()
```
